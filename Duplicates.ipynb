```python
import pandas as pd
```

```python
df=pd.DataFrame({'col1':['A','A','B','B','B'], 'COL2':[2,2,2,3,3]})
```

```python
df
```

```python
df.duplicated()
```

```python
df1=df.drop_duplicates()
```

```python
df1
```

```python
df2=df1.reset_index()
```

```python
df2
```

```python
df3=df2.drop(columns='index', axis=1)
```

```python
df3
```
